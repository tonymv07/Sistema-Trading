```python
import requests
import json 
import time
from bs4 import BeautifulSoup


class main_coin():
        def get_current_price_btc(self):
            url = "https://api.coingecko.com/api/v3/coins/bitcoin"
            payload = {}
            headers = {}
            response = requests.request("GET", url, headers=headers, data=payload)
            USD_coin = float(response.json()["market_data"]["current_price"]["usd"])
            return USD_coin
        def main(self):
            datos = requests.get('https://www.coingecko.com/es/monedas/recently_added').content
            soup = BeautifulSoup(datos, 'html.parser')
            data_all_coins = []
            value_btc_usd = self.get_current_price_btc()
            for a_tag in soup.findAll("i"):
                symbol = a_tag.attrs.get('data-coin-symbol')
                id = a_tag.attrs.get('data-coin-id')
                name = a_tag.attrs.get('data-coin-slug')
                price_btc = a_tag.attrs.get('data-price-btc')
                last_added = a_tag.attrs.get('trade p-0 col-market pl-2 text-center')
                data_coin = {}
                data_coin["symbol"] = ""
                data_coin["id"] = ""
                data_coin["name"] = ""
                data_coin["price_btc"] = "" 
                data_coin["last_added"] = ""
                if symbol != None and  id != None and name != None and price_btc != None:
                        data_coin["symbol"] = symbol
                        data_coin["id"] = id
                        data_coin["name"] = name
                        data_coin["price_btc"] = "{:.15f}".format(float(price_btc) * value_btc_usd)
                        data_all_coins.append(data_coin)
            i = 0
            for a_tag in soup.findAll("td", {'class':'trade p-0 col-market pl-2 text-center'}): 
                data_all_coins[i]["last_added"] = a_tag.get_text('self.get_current_price_btc').replace("\n", "")
                i = i + 1

            for obj in data_all_coins:
                print(obj)
                
main_coin().main()
def new_coins():
    list_coins="main_coin().main()"
    return list_coins
while new_coins == True:
    time.sleep(0.5)
```
